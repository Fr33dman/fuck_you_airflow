```python
from pandas import read_excel

from sberpm.ml.text_clustering._wrapper_text_clustering import WrapperTextClustering
```

```python
df = read_excel('example.xlsx')
df.head()
```

#### TextClustering with default parameters and example data

```python
text_clust = WrapperTextClustering(
    data=df,
    model_params={
        "notation_params": dict(
            id_col="id",
            status_col="stages",
            date_col="dt",
            date_end_col=None,
            user_id_col=None,
            other="some_text",
        ),
        "model_params": dict(
            description="some_text",
            pca_dim="full_quality",
            type_model_w2v="navec",
            min_samples=5,
            only_unique_descriptions=False,
            cluster_marking=[],
        ),
    },
)
```

```python
%%time

text_clustering_output = text_clust.run_model()
```
